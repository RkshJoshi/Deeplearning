## Tensorflow fundamental exercise

```python
import numpy as np
import tensorflow as tf
```

```python
# Defining a vector
ten1 = tf.constant([1,2,3])
ten1.ndim
```

```python
ten2 = tf.constant(5)
ten2.ndim   # this is a scaler tensor
```

```python
ten3 = tf.constant([[2,5],[6,7]])
ten3.ndim
```

```python
ten4 = tf.constant([[[2,5,6],
                    [4,6,7]],
                   [[5,6,7],
                   [6,7,9]]])
ten4.ndim
```

```python
ten4.shape, tf.rank(ten4), tf.size(ten4)
```

```python
t5 = tf.random.uniform(shape=(5,300), minval=0, maxval=1, dtype=tf.float16)
t6 = tf.random.uniform(shape=(5,300), minval=0, maxval=1, dtype=tf.float16)
t5, t6
```

```python
t5 * t6
```

```python
tf.matmul(t5, tf.transpose(t6))
```

```python
t6 = tf.random.uniform(shape=(224,224,3), minval=0, maxval=1, dtype=tf.float16)
t6
```

```python
tf.reduce_max(t6, axis=1), tf.reduce_min(t6, axis=1)
```
